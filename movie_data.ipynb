```python
wiki_movies = [movie for movie in wiki_movies_raw
                if('Director' in movie or 'Directed by' in movie) 
                    and 'imdb_link' in movie
                    and 'No. of episodes' not in movie]
len(wiki_movies)
```

```python
wiki_movies2_df = pd.DataFrame(wiki_movies)
wiki_movies2_df.head()
```

```python
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    alt_title_keys = ['Arabic',
                        'Also known as',
                        'Cantonese',
                        'Chinese',
                        'French',
                        'Hangul',
                        'Hebrew',
                        'Hepburn',
                        'Japanese',
                        'Literally',
                        'Mandarin',
                        'McCune–Reischauer',
                        'Original title',
                        'Polish',
                        'Revised Romanization',
                        'Romanized',
                        'Russian',
                        'Simplified',
                        'Traditional',
                        'Yiddish']
    for key in alt_title_keys:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles)>0:
        movie['alt_titles']= alt_titles
            
    
    return movie
```

```python
clean_movies = [clean_movie(movie) for movie in wiki_movies]
wiki_movies3_df = pd.DataFrame(clean_movies)
sorted(wiki_movies3_df.columns.to_list())
```
